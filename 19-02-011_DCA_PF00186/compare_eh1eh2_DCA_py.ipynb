```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ehDCA = np.loadtxt('result_MU.dat').T
```

```python
ehDCA.shape
```

```python
eh = np.loadtxt('eh1eh2.dat').T
```

```python
eh.shape
```

```python
eh1DCA,eh2DCA = ehDCA[3,:],ehDCA[4,:]
eh12DCA = eh1DCA+eh2DCA

eh1,eh2 = eh[3,:],eh[4,:]
eh12 = eh1+eh2

#--------------------------------
plt.figure(figsize=(11,3.2))

plt.subplot2grid((1,3),(0,0))
plt.title('eh1')
plt.plot([eh1.min(),eh1.max()],[eh1.min(),eh1.max()],'r--')
plt.scatter(eh1DCA,eh1)

plt.subplot2grid((1,3),(0,1))
plt.title('eh2')
plt.plot([eh2.min(),eh2.max()],[eh2.min(),eh2.max()],'r--')
plt.scatter(eh2DCA,eh2)

plt.subplot2grid((1,3),(0,2))
plt.title('eh1+eh2')
plt.plot([eh12.min(),eh12.max()],[eh12.min(),eh12.max()],'r--')
plt.scatter(eh12DCA,eh12)

plt.tight_layout(h_pad=1, w_pad=1.5)
plt.savefig('eh1eh2_compare.pdf', format='pdf', dpi=100)
plt.show()
```
